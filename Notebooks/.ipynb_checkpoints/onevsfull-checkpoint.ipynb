```python
MMFP_time = []
MMFP_gini = []
ONE_time = []
ONE_gini = []

for orig_node, dest_node in tqdm(zip(st[0], st[1]), total=len(st[0]), desc="Processing pairs of nodes"):

    start_time = time.time()
    dag = ut.get_dag(g, orig_node, dest_node, weight=wgt)
    dag_time = time.time() - start_time
    
    K, alpha, model, result = ut.iterative_solver(dag, orig_node, dest_node) #LP solver, alpha are the satisf. probs
    end_time = time.time()
    MMFP_gini.append(ut.get_gini_from_alpha(alpha))    
    MMFP_time.append(end_time - start_time)
    
    start_time = time.time()
    result = ut.max_flow_with_alpha(dag, orig_node, dest_node)
    final = ut.compute_probabilities_and_expectations(dag, result, dest_node) 
    dag = final['dag']
    ut.compute_and_assign_node_flow(dag, dag, result['flow'],orig_node)
    ONE_gini.append(ut.get_gini(dag,'node_flow'))
    end_time = time.time()
    ONE_time.append(end_time - start_time + dag_time)
```

```python
MMFP_time = []
MMFP_gini = []
ONE_time = []
ONE_gini = []

for orig_node, dest_node in tqdm(zip(st[0], st[1]), total=len(st[0]), desc="Processing pairs of nodes"):

    start_time = time.time()
    dag = ut.get_dag(g, orig_node, dest_node, weight=wgt)
    dag1 = copy.deepcopy(dag)
    dag_time = time.time() - start_time
    
    
    K, alpha, model, result = ut.iterative_solver(dag, orig_node, dest_node) #LP solver, alpha are the satisf. probs
    final = ut.compute_probabilities_and_expectations(dag, result, dest_node) 
    dag = final['dag']
    ut.compute_and_assign_node_flow(dag, dag, result['flow'],orig_node)    
    end_time = time.time()
    MMFP_gini.append(ut.get_gini(dag,'node_flow'))    
    MMFP_time.append(end_time - start_time)
    
    
    start_time = time.time()
    result = ut.max_flow_with_alpha(dag1, orig_node, dest_node)
    final = ut.compute_probabilities_and_expectations(dag1, result, dest_node) 
    dag1 = final['dag']
    ut.compute_and_assign_node_flow(dag1, dag1, result['flow'],orig_node)
    ONE_gini.append(ut.get_gini(dag1,'node_flow'))
    end_time = time.time()
    ONE_time.append(end_time - start_time + dag_time)
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(ONE_time, ONE_gini, color='red', label='ONE', alpha=0.6)
plt.scatter(MMFP_time, MMFP_gini, color='blue', label='MMFP', alpha=0.6)


# Adding labels and title
plt.xlabel('Time (seconds)')
plt.ylabel('Gini Coefficient')
plt.title('Scatter Plot of Time vs Gini Coefficient for MMFP and ONE Methods')
plt.legend(loc = 'lower right')

# Show the plot

plt.show()
```

```python
def subtract_values(set_a, set_b):
    result = {}

    # Convert set_b to a dictionary for easier lookup
    dict_b = dict(set_b)
    
    for key, value in set_a.items():
        if key in dict_b:
            result[key] = value - dict_b[key]
        else:
            result[key] = value
    
    # Check for keys in set_b that are not in set_a
    for key, value in dict_b.items():
        if key not in result:
            result[key] = -value
    
    return result
subtract_values(alpha, set(dag.nodes(data = 'node_flow')))
```
