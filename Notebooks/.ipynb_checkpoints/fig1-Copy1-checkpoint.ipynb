```python
name = "bison"
g = gt.collection.ns[name]
print(g.edge_properties)
```

```python
if not g.edge_properties:
    weights = g.new_edge_property("double")
    weights.set_value(1.0)  # Set all values to 1.0
    g.edge_properties["weight"] = weights
    print(g.edge_properties)
```

```python
if "weight" in g.edge_properties:
    layer_property = g.edge_properties["weight"]
    for i,e in enumerate(g.edges()):
        if i >= 5:
            break
        print("Edge:", e, "width:", layer_property[e])
```

```python
def from_gt_to_nx(g, wgt = "distance"):
    g_nx = nx.DiGraph()
    
    for v in g.vertices():
        g_nx.add_node(int(v))
    
    for e in g.edges():
        source = int(e.source())
        target = int(e.target())
        weight = g.ep[wgt][e]  
        g_nx.add_edge(source, target, weight=weight)
        #g_nx.add_edge(target, source, weight=weight) #if undirected graph
    return(g_nx)
```

```python
g_nx = from_gt_to_nx(g, wgt = "weight")
weights = g.edge_properties["weight"]
```

```python
source = 0
target = 12
dist, pred = gt.search.dijkstra_search(g, source = source, weight=weights)
temp = dist[target]

if temp < np.inf:
    shortest_path = []
    current = target
    while current != source:
        shortest_path.append(current)
        current = pred[current]
    shortest_path.append(source)
    shortest_path.reverse()
print(shortest_path)
```

```python
plt.figure(figsize=(8, 6))
pos = nx.kamada_kawai_layout(g_nx)  # Position nodes using Fruchterman-Reingold force-directed algorithm
nx.draw(g_nx, pos, with_labels=True, node_size=3000, node_color='lightblue', font_size=12, font_weight='bold', arrowsize=20)
plt.title('Directed Graph with Node Names')
plt.show()
```

```python
import matplotlib.pyplot as plt
import networkx as nx

def custom_kamada_kawai_layout(G, source, target):
    # Use Kamada-Kawai layout for initial positioning
    pos = nx.kamada_kawai_layout(G)
    
    # Adjust positions for the source and target nodes
    min_x = min(pos[source][0] for source in G.nodes)
    max_x = max(pos[target][0] for target in G.nodes)
    
    pos[source] = (min_x - 1, pos[source][1])  # Place source node on the left
    pos[target] = (max_x + 1, pos[target][1])  # Place target node on the right
    
    return pos


# Generate the custom layout
pos = custom_kamada_kawai_layout(g_nx, source, target)

# Draw the graph with the custom layout
plt.figure(figsize=(10, 6))
nx.draw(g_nx, pos, with_labels=True, node_size=3000, node_color='lightblue', font_size=12, font_weight='bold', arrowsize=20)
plt.title('Custom Kamada-Kawai Layout with Source on the Left and Target on the Right')
plt.show()

```

```python
dist_dict = nx.shortest_path_length(g_nx, source=None, target=target, weight='weight', method='dijkstra')
```

```python
def remove_edges_based_on_scores(graph, dist_dict):
    edges_to_remove = []
    for node in graph.nodes():
        try:
            node_score = dist_dict[node]
            for neighbor in graph.neighbors(node):
                try:
                    neighbor_score = dist_dict[neighbor]
                except KeyError: 
                    neighbor_score = np.inf
                if neighbor_score >= node_score:
                    edges_to_remove.append((node, neighbor))
        except KeyError:
            continue
    for edge in edges_to_remove:
        graph.remove_edge(*edge)

G_copy = copy.deepcopy(g_nx)

remove_edges_based_on_scores(G_copy, dist_dict)
```

```python
def remove_unreachable_nodes(G, source):
    # Get the set of nodes reachable from the source
    reachable_nodes = nx.descendants(G, source) | {source}


    # Find nodes that are not reachable
    non_reachable_nodes = set(G.nodes()) - reachable_nodes

    # Remove non-reachable nodes from the graph
    G.remove_nodes_from(non_reachable_nodes)
    
    return G
remove_unreachable_nodes(G_copy, source)
G_copy.edges()
```
