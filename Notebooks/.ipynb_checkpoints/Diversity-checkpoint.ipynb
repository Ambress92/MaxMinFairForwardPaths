```python
def filter_st(g, wgt, st):

    # Resultant lists after filtering
    filtered_sources = []
    filtered_destinations = []
    
    for src, dest in zip(st[0], st[1]):
        try:
            # Compute the shortest path using Dijkstra's algorithm
            shortest_path = nx.dijkstra_path(g, src, dest, weight=wgt)
            # Check the number of nodes in the shortest path
            if len(shortest_path) > 6:
                filtered_sources.append(src)
                filtered_destinations.append(dest)
        except nx.NetworkXNoPath:
            # If there is no path between src and dest
            continue
    
    return [filtered_sources, filtered_destinations]
```

```python
import numpy as np
from contextlib import contextmanager
import sys
import os
import pandas as pd
import matplotlib.pyplot as plt
from tqdm import tqdm  # Import tqdm
import time  # Import time module

# Context manager to suppress prints
@contextmanager
def suppress_output():
    with open(os.devnull, 'w') as devnull:
        old_stdout = sys.stdout
        old_stderr = sys.stderr
        try:
            sys.stdout = devnull
            sys.stderr = devnull
            yield
        finally:
            sys.stdout = old_stdout
            sys.stderr = old_stderr

# Initialize lists to store the results and runtimes
results_FP_paths = []
results_random_FP_paths = []
results_yen_k_shortest = []
results_kspwlo_paths_08 = []
results_kspwlo_paths_05 = []

runtimes_FP_paths = []
runtimes_random_FP_paths = []
runtimes_yen_k_shortest = []
runtimes_kspwlo_paths_08 = []
runtimes_kspwlo_paths_05 = []

# Iterate through pairs of nodes with a progress bar
for orig_node, dest_node in tqdm(zip(st[0], st[1]), total=len(st[0]), desc="Processing pairs of nodes"):
    with suppress_output():
        # Measure time for the first block
        start_time = time.time()
        dag = ut.get_dag(g, orig_node, dest_node, weight=wgt)
        result = ut.max_flow_with_alpha(dag, orig_node, dest_node)
        final = ut.compute_probabilities_and_expectations(dag, result, dest_node)
        dag = final['dag']
        mid time =time.time()
        elapsed_dag_time = mid_time - start_time
        
        start_time = time.time()
        FP_paths = ut.sample_k_FP(dag, orig_node, dest_node, k)
        end_time = time.time()
        elapsed_time_FP_paths = end_time - start_time + elapsed_dag_time
        runtimes_FP_paths.append(elapsed_time_FP_paths)

        #Measure time for random_FP_paths
        start_time = time.time()
        random_FP_paths = ut.sample_k_random_FP(dag, orig_node, dest_node, k)
        end_time = time.time()
        elapsed_time_random_FP = end_time - start_time + elapsed_dag_time
        runtimes_random_FP_paths.append(elapsed_time_random_FP)

        # Measure time for Yen's K-shortest paths
        start_time = time.time()
        yen_k_shortest = ut.yen_k_paths(g, orig_node, dest_node, weight=wgt, k=10)
        end_time = time.time()
        elapsed_time_yen = end_time - start_time
        runtimes_yen_k_shortest.append(elapsed_time_yen)

        # Measure time for KSPwlo 0.8
        start_time = time.time()
        node_map = ut.save_graph_to_gr(g, 'tmp.gr', source=orig_node, target=dest_node, weight=wgt)
        kspwlo_paths_08 = ut.inverse_mapping(
            ut.execute_bash_command("tmp.gr", k, 0.8, node_map[orig_node], node_map[dest_node], "opplus"), node_map)
        end_time = time.time()
        elapsed_time_kspwlo_08 = end_time - start_time
        runtimes_kspwlo_paths_08.append(elapsed_time_kspwlo_08)

        # Measure time for KSPwlo 0.9
        start_time = time.time()
        kspwlo_paths_05 = ut.inverse_mapping(
            ut.execute_bash_command("tmp.gr", k, 0.9, node_map[orig_node], node_map[dest_node], "opplus"), node_map)
        end_time = time.time()
        elapsed_time_kspwlo_05 = end_time - start_time
        runtimes_kspwlo_paths_05.append(elapsed_time_kspwlo_05)

    for paths, results in zip([FP_paths, random_FP_paths, yen_k_shortest, kspwlo_paths_05, kspwlo_paths_08],
                              [results_FP_paths, results_random_FP_paths, results_yen_k_shortest, results_kspwlo_paths_05, results_kspwlo_paths_08]):
        mean_norm_sym_diff = ut.mean_normalized_symmetric_difference_nodes(paths)
        avg_path_length = ut.average_path_length(paths, g, wgt)
        results.append((mean_norm_sym_diff, avg_path_length))

# Function to calculate mean and standard deviation for each metric
def calculate_mean_std(results):
    mean_norm_sym_diff_list = [r[0] for r in results]
    avg_path_length_list = [r[1] for r in results]
    mean_norm_sym_diff_mean = np.mean(mean_norm_sym_diff_list)
    mean_norm_sym_diff_std = np.std(mean_norm_sym_diff_list)
    avg_path_length_mean = np.mean(avg_path_length_list)
    avg_path_length_std = np.std(avg_path_length_list)
    return mean_norm_sym_diff_mean, mean_norm_sym_diff_std, avg_path_length_mean, avg_path_length_std

# Function to calculate mean and standard deviation for runtimes
def calculate_runtime_mean_std(runtimes):
    runtime_mean = np.mean(runtimes)
    runtime_std = np.std(runtimes)
    return runtime_mean, runtime_std

# Calculate mean and standard deviation for each method
methods = ["FP_paths", "random_FP_paths", "yen_k_shortest", "kspwlo_paths_05", "kspwlo_paths_08"]
results_lists = [results_FP_paths, results_random_FP_paths, results_yen_k_shortest, results_kspwlo_paths_05, results_kspwlo_paths_08]
runtimes_lists = [runtimes_FP_paths, runtimes_random_FP_paths, runtimes_yen_k_shortest, runtimes_kspwlo_paths_05, runtimes_kspwlo_paths_08]

data = []

for method, results, runtimes in zip(methods, results_lists, runtimes_lists):
    mean_norm_sym_diff_mean, mean_norm_sym_diff_std, avg_path_length_mean, avg_path_length_std = calculate_mean_std(results)
    runtime_mean, runtime_std = calculate_runtime_mean_std(runtimes)
    data.append([method, mean_norm_sym_diff_mean, mean_norm_sym_diff_std, avg_path_length_mean, avg_path_length_std, runtime_mean, runtime_std])

# Create a pandas DataFrame
df = pd.DataFrame(data, columns=["Method", "Diversity (Mean)", "Diversity (Std)", "Avg Path Length (Mean)", "Avg Path Length (Std)", "Runtime (Mean)", "Runtime (Std)"])

# Print the DataFrame
print(df)
```

```python
# Extract data for plotting
x = df['Avg Path Length (Mean)']
x_err = df['Avg Path Length (Std)']
y = df['Diversity (Mean)']
y_err = df['Diversity (Std)']

# Create the plot
plt.figure(figsize=(10, 6))
plt.errorbar(x, y, xerr=x_err, yerr=y_err, fmt='o', capsize=5, label='Methods')

# Annotate each point with the corresponding method name
for i, method in enumerate(df['Method']):
    plt.annotate(method, (x[i], y[i]), textcoords="offset points", xytext=(5,5), ha='center')

# Labels and title
plt.xlabel('Average Path Length (Mean)')
plt.ylabel('Diversity (Mean)')
plt.title('Method Comparison: Diversity vs. Average Path Length')
plt.grid(True)
plt.legend()

# Reverse x-axis
plt.gca().invert_xaxis()

# Show plot
plt.show()
```
