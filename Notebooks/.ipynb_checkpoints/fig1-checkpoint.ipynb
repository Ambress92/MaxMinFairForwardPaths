```python
def from_gt_to_nx(g, wgt = "distance"):
    g_nx = nx.DiGraph()
    
    for v in g.vertices():
        g_nx.add_node(int(v))
    
    for e in g.edges():
        source = int(e.source())
        target = int(e.target())
        weight = g.ep[wgt][e]  
        g_nx.add_edge(source, target, weight=weight)
        #g_nx.add_edge(target, source, weight=weight) #if undirected graph
    return(g_nx)
```

```python
g_nx = from_gt_to_nx(g, wgt = "weight")
weights = g.edge_properties["weight"]
```

```python
dist_dict = nx.shortest_path_length(g_nx, source=None, target=target, weight='weight', method='dijkstra')
dist_dict
```

```python
def remove_edges_based_on_scores(graph, dist_dict):
    edges_to_remove = []
    for node in graph.nodes():
        try:
            node_score = dist_dict[node]
            for neighbor in graph.neighbors(node):
                try:
                    neighbor_score = dist_dict[neighbor]
                except KeyError: 
                    neighbor_score = np.inf
                if neighbor_score >= node_score:
                    edges_to_remove.append((node, neighbor))
        except KeyError:
            continue
    for edge in edges_to_remove:
        graph.remove_edge(*edge)

G_copy = copy.deepcopy(g_nx)

remove_edges_based_on_scores(G_copy, dist_dict)
```

```python
def remove_unreachable_nodes(G, source):
    # Get the set of nodes reachable from the source
    reachable_nodes = nx.descendants(G, source) | {source}


    # Find nodes that are not reachable
    non_reachable_nodes = set(G.nodes()) - reachable_nodes

    # Remove non-reachable nodes from the graph
    G.remove_nodes_from(non_reachable_nodes)
    
    return G
remove_unreachable_nodes(G_copy, source)
G_copy.edges()
```

```python
import matplotlib.pyplot as plt
import networkx as nx

def custom_kamada_kawai_layout(G, source, target):
    # Use Kamada-Kawai layout for initial positioning
    pos = nx.kamada_kawai_layout(G)
    
    # Adjust positions for the source and target nodes
    min_x = min(pos[node][0] for node in G.nodes)
    max_x = max(pos[node][0] for node in G.nodes)
    
    pos[source] = (min_x - 1, pos[source][1])  # Place source node on the left
    pos[target] = (max_x + 1, pos[target][1])  # Place target node on the right
    
    return pos

# Compute all shortest paths between source and target
shortest_paths = list(nx.all_shortest_paths(g_nx, source=source, target=target, weight='weight'))
print(shortest_paths)

# Generate the custom layout
pos = custom_kamada_kawai_layout(g_nx, source, target)

# Draw the graph with the custom layout
plt.figure(figsize=(10, 6))

# Color nodes in G_copy red
node_color = ['orange' if any(node in path for path in shortest_paths) else 'red' if node in G_copy.nodes() else 'lightblue' for node in g_nx.nodes()]

nx.draw(g_nx, pos, with_labels=True, node_size=1000, node_color=node_color, font_size=12, font_weight='bold', arrowsize=20)

# Extract edge labels (weights) for displaying
edge_labels = {(u, v): g_nx[u][v]['weight'] for u, v in g_nx.edges()}

# Color edges and nodes of the shortest paths in orange
for path in shortest_paths:
    for u, v in zip(path[:-1], path[1:]):
        if (u, v) in g_nx.edges():
            nx.draw_networkx_edges(g_nx, pos, edgelist=[(u, v)], edge_color='orange', width=2.0, arrowsize=20)
            # Add edge labels with orange color
            nx.draw_networkx_edge_labels(g_nx, pos, edge_labels={(u, v): edge_labels[(u, v)]}, font_color='orange')

for u, v in g_nx.edges():
    if (u, v) in G_copy.edges() and not any((u, v) in zip(path, path[1:]) for path in shortest_paths):
        nx.draw_networkx_edges(g_nx, pos, edgelist=[(u, v)], edge_color='red', width=2.0, arrowsize=20)
        # Add edge labels with red color
        nx.draw_networkx_edge_labels(g_nx, pos, edge_labels={(u, v): edge_labels[(u, v)]}, font_color='red')

plt.title('Custom Kamada-Kawai Layout with Source on the Left and Target on the Right')
plt.show()
```

```python
import matplotlib.pyplot as plt
import networkx as nx

def custom_kamada_kawai_layout(G, source, target):
    # Use Kamada-Kawai layout for initial positioning
    pos = nx.kamada_kawai_layout(G)
    
    # Adjust positions for the source and target nodes
    min_x = min(pos[node][0] for node in G.nodes)
    max_x = max(pos[node][0] for node in G.nodes)
    
    pos[source] = (min_x - 1, pos[source][1])  # Place source node on the left
    pos[target] = (max_x + 1, pos[target][1])  # Place target node on the right
    
    return pos

# Compute all shortest paths between source and target
shortest_paths = list(nx.all_shortest_paths(g_nx, source=source, target=target, weight = 'weight'))
print(shortest_paths)
# Generate the custom layout
pos = custom_kamada_kawai_layout(g_nx, source, target)

# Draw the graph with the custom layout
plt.figure(figsize=(10, 6))

# Color nodes in G_copy red
node_color = ['orange' if any(node in path for path in shortest_paths) else 'red' if node in G_copy.nodes() else 'lightblue' for node in g_nx.nodes()]

nx.draw(g_nx, pos, with_labels=True, node_size=1000, node_color=node_color, font_size=12, font_weight='bold', arrowsize=20)

# Extract edge labels (weights) for displaying
edge_labels = {(u, v): f'{d["weight"]}' for u, v, d in g_nx.edges(data=True)}

# Color edges and nodes of the shortest paths in orange
for path in shortest_paths:
    for u, v in zip(path[:-1], path[1:]):
        if (u, v) in g_nx.edges():
            nx.draw_networkx_edges(g_nx, pos, edgelist=[(u, v)], edge_color='orange', width=2.0, arrowsize=20)
            # Add edge labels with orange color
            nx.draw_networkx_edge_labels(g_nx, pos, edge_labels={(u, v): f'{d["weight"]}'}, font_color='orange')

for u, v, d in g_nx.edges(data=True):
    if (u, v) in G_copy.edges() and not any((u, v) in zip(path, path[1:]) for path in shortest_paths):
        nx.draw_networkx_edges(g_nx, pos, edgelist=[(u, v)], edge_color='red', width=2.0, arrowsize=20)
        # Add edge labels with red color
        nx.draw_networkx_edge_labels(g_nx, pos, edge_labels={(u, v): f'{d["weight"]}'}, font_color='red')

plt.title('Custom Kamada-Kawai Layout with Source on the Left and Target on the Right')
plt.show()
```
