```python
from math import exp, log
import igraph as ig
import matplotlib.pyplot as plt
import numpy as np
import gurobipy as gp
from gurobipy import GRB


def build_weighted_igraph(vertices, weighted_edges):
    G = ig.Graph(directed=True)
    G.add_vertices(vertices)
    G.add_edges([(e[0], e[1]) for e in weighted_edges])
    G.es['weight'] = [e[2] for e in weighted_edges]
    return G

def visualize(graph, label="weight"):
    layout = graph.layout("fr")
    fig, ax = plt.subplots(figsize=(7, 7))
    visual_style = {
        "vertex_label": graph.vs["name"],
        "edge_label": graph.es[label],
        "layout": layout,
        "bbox": (300, 300),
        "margin": 20
    }
    ig.plot(graph, target=ax, **visual_style)
    plt.show()

# Create a forward DAG from s to t and also orders its vertices in order of decreasing distances to t
def create_forward_DAG(graph, s, t):
    dist_from_s = graph.distances(source=s, mode="out", weights='weight')[0]
    dist_to_t = graph.distances(source=t, mode="in", weights='weight')[0]

    reachable_from_s = {i for i, dist in enumerate(dist_from_s) if dist < float('inf')}
    reachable_to_t = {i for i, dist in enumerate(dist_to_t) if dist < float('inf')}
    reachable = reachable_from_s.intersection(reachable_to_t)
    reachable = sorted(reachable, key=lambda i:-dist_to_t[i])
    subgraph = graph.induced_subgraph(reachable)

    # Identify the edges where the distance from u to t is larger than the distance from v to t
    valid_edges = []
    valid_weights = []
    for edge in subgraph.es:
        u, v = edge.source, edge.target
        if dist_to_t[u] > dist_to_t[v]:
            valid_edges.append((subgraph.vs[u]["name"], subgraph.vs[v]["name"]))
            valid_weights.append(edge["weight"])

    # Create a new graph with only the valid edges and their weights
    new_graph = ig.Graph(directed=True)
    new_graph.add_vertices(subgraph.vs[reachable]["name"])
    new_graph.add_edges(valid_edges)
    new_graph.es['weight'] = valid_weights
    return new_graph

def solve_fairness_model(dag):
    s, t = dag.vs[0], dag.vs[-1]
    model = gp.Model()

    # Add variables: alpha and flow (one per edge)
    alpha = model.addVar(lb=0)
    f = model.addVars(dag.es, lb=0)
    model.update()

    # Constraints: prob visiting source and target
    model.addConstr( gp.quicksum(f[e] for e in s.out_edges()) == 1 )
    model.addConstr( gp.quicksum(f[e] for e in t.in_edges()) == 1 )

    # Constraints: out flow = in flow for every v != s, t
    flow_cons = model.addConstrs( ( gp.quicksum(f[e] for e in v.out_edges()) == gp.quicksum(f[e] for e in v.in_edges()) ) for v in dag.vs[1:-1] )
        
    # Constraints: in flow >= alpha for every v != s, t
    alpha_cons = model.addConstrs( ( gp.quicksum(f[e] for e in v.in_edges()) >= alpha ) for v in dag.vs[1:-1] )
    model.update()

    # Objective function
    model.setObjective(alpha, GRB.MAXIMIZE)

    # Optimize the model
    model.optimize()
    return model, f, (flow_cons, alpha_cons)


# Analyze the solution, modify the dag to include also probabilities and conditional probabilities, and return some statistics
def analyze_solution(dag, model, all_vars, all_cons):    
    fairness = model.objVal
    f = all_vars
    flow_cons, alpha_cons = all_cons

    # Get the dual weights
    w = np.zeros(dag.vcount())
    d = np.zeros(dag.vcount())
    for (v, cons) in alpha_cons.items():
        w[v.index] = -cons.Pi
    for (v, cons) in flow_cons.items():
        d[v.index] = -cons.Pi
    d[-1] = fairness    
    assert abs(w.sum() - 1) < 1e-6

    # Add the flow to the DAG as probabilities and compute also conditional probabilities
    for v in dag.vs:
        # Calculate total outgoing flow for each vertex
        total_out_flow = sum(f[e].X for e in v.out_edges())
        v['prob'] = total_out_flow
        v['dual_weight'] = d[v.index]
        
        for e in v.out_edges():
            flow_value = f[e].X  # Get the optimal flow value for this edge
            e['prob'] = flow_value  # Add 'prob' attribute to the edge
            
            # Calculate and add conditional probability
            if total_out_flow > 0:
                e['cond_prob'] = flow_value / total_out_flow
            else:
                e['cond_prob'] = 0 
    dag.vs[-1]['prob'] = 1                

    # Compute the expected length (edge weights) of a path
    exp_length = (np.array(dag.es['weight'])*np.array(dag.es['prob'])).sum()

    # Compute the expected number of vertices visited in a path (including s and t)
    exp_numv = 1 + np.array(dag.es['prob']).sum()
    return {"fairness":fairness, "exp_length":exp_length, "exp_numv":exp_numv}


def input_graph():
    vertices = ["A", "B", "C", "D", "E"]
    weighted_edges = [("E", "A", 4), ("E", "B", 3), ("B", "C", 2), ("C", "D", 1), ("A", "D", 1), ("A", "B", 1), ("C", "A", 1)]

    return build_weighted_igraph(vertices, weighted_edges)

def check_dag(dag):
    assert [ v.index for v in dag.vs ] == list( range( dag.vcount() ) )
    assert all( e.source < e.target for e in dag.es )

graph = input_graph()    
dag = create_forward_DAG(graph, "E", "D")
#visualize(dag)
check_dag(dag)

model, all_vars, all_cons = solve_fairness_model(dag)    
print( analyze_solution(dag, model, all_vars, all_cons) )

# Print the probabilities for each edge
print("Edge Probabilities:")
for e in dag.es:
    source = dag.vs[e.source]['name']  # Assuming vertices have a 'name' attribute
    target = dag.vs[e.target]['name']
    print(f"Edge ({source} -> {target}): prob = {e['prob']:.4f}, cond_prob = {e['cond_prob']:.4f}")
visualize(dag, 'cond_prob')

print("Vertex dual weights:")
for v in dag.vs:
    print(v['name'], v['dual_weight'])
```
